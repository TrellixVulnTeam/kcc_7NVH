```python
import pandas as pd
from nltk.translate.bleu_score import sentence_bleu
```

## TST train

```python
target_df = pd.read_fwf("../../kcc_data/tst_train_target_decode.txt", header=None)
output_df = pd.read_fwf("../../kcc_data/tst_train_output_decode.txt", header=None)

target_df["output"] = output_df
target_df.columns = ["target", "output"]
target_df
```

```python
reference = []
candidate = []
reference.append([i.split() for i in target_df.target.tolist()])
candidate.append([i.split() for i in target_df.output.tolist()])
reference = reference[0]
candidate = candidate[0]

total_bleu_1gram = 0
total_bleu_2gram = 0
total_bleu_3gram = 0
total_bleu_4gram = 0

for i in range(len(target_df)):
    print(f"target: {target_df.target[i]},\noutput: {target_df.output[i]}")
    bleu_1gram = sentence_bleu([reference[i]], candidate[i], weights=(1, 0, 0, 0))
    bleu_2gram = sentence_bleu([reference[i]], candidate[i], weights=(0, 1, 0, 0))
    bleu_3gram = sentence_bleu([reference[i]], candidate[i], weights=(0, 0, 1, 0))
    bleu_4gram = sentence_bleu([reference[i]], candidate[i], weights=(0, 0, 0, 1))
    total_bleu_1gram += bleu_1gram
    total_bleu_2gram += bleu_2gram
    total_bleu_3gram += bleu_3gram
    total_bleu_4gram += bleu_4gram
    
#     print(f'reference: {reference[i]}')
#     print(f'candidate: {candidate[i]}\n')
    print(f'1-Gram BLEU: {bleu_1gram:.2f}')
    print(f'2-Gram BLEU: {bleu_2gram:.2f}')
    print(f'3-Gram BLEU: {bleu_3gram:.2f}')
    print(f'4-Gram BLEU: {bleu_4gram:.2f}')
    print("===================================")

print(f'Avg 1-Gram BLEU: {total_bleu_1gram/len(target_df):.2f}')
print(f'Avg 2-Gram BLEU: {total_bleu_2gram/len(target_df):.2f}')
print(f'Avg 3-Gram BLEU: {total_bleu_3gram/len(target_df):.2f}')
print(f'Avg 4-Gram BLEU: {total_bleu_4gram/len(target_df):.2f}')
```
