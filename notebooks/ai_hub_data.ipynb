```python
import pandas as pd
```

```python
news = pd.read_excel("../../kor_dataset/3_문어체_뉴스(1)_200226.xlsx", header=None)
cdf = pd.read_excel("../../kor_dataset/2_대화체.xlsx", header=None)
print(news.shape, cdf.shape)
news.head()
```

```python
news_df = news.iloc[1:, 7:]
conv_df = cdf.iloc[1:, 5:]
print(news_df.shape, conv_df.shape)
news_df.head()
```

```python
conv_df.head()
```

```python
news_df.columns = ["kor", "eng"]
conv_df.columns = ["kor", "eng"]
news_df = news_df[["eng", "kor"]]
conv_df = conv_df[["eng", "kor"]]
print(news_df.shape, conv_df.shape)
news_df.head()
```

```python
conv_df.head()
```

```python
total_df = pd.concat([news_df, conv_df], axis=0)
print(total_df.shape)
total_df.head()
```

```python
with open ("../.data/ai_hub/total_eng.txt", "w", encoding="utf-8") as ef:
    for line in total_df["eng"]:
        ef.write(f"{line}\n")
with open("../.data/ai_hub/total_kor.txt", "w", encoding="utf-8") as kf:
    for line in total_df["kor"]:
        kf.write(f"{line}\n")
    
    
```
